```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

# Importing the dataset


```python
dataset = pd.read_csv('50_Startups.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 4].values

print(X.shape, y.shape)
```

# Encoding the Independent Variable (State)

```python
# Encoding the Independent variable (dummy encoding)
from sklearn.preprocessing import OneHotEncoder

onehotencoder_X1 = OneHotEncoder()
X1 = X[:, 3].reshape(-1, 1)#make sure x1 is 2 dimension, row = -1 means don't care. choose the row3 (state col) 
onehotX1_result = onehotencoder_X1.fit_transform(X1).toarray()
#onehotencoder_X1.categories_
# print(onehotX1_result)

X = np.c_[X[:, 0:3],onehotX1_result] # c_ means combine., we throw the first column
print(X) # now X state is change to the one hot encoder data. 
```

# Avoiding the Dummy Variable Trap


```python
X = X[:, :-1] #去掉最後一個data，因為前兩者皆為0時，代表最後一個必為1
print(X[0:5,:])
```


# Splitting the dataset into the Training set and Test set


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)#40筆資料拿去做訓練，10筆資料拿來做測試
print(X_train[0:5,:])
```

# Fitting Multiple Linear Regression to the Training set


```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

# Predicting the Test set results


```python
y_pred = regressor.predict(X_test)
print('coef= ', regressor.coef_)
print('intercept= ', regressor.intercept_)
print('score= ', regressor.score(X, y))
```

# Building the optimal model using Backward Elimination






```python
import statsmodels.api as sm
# X_train = np.append(arr = np.ones((40, 1)), values = X_train, axis = 1)
X_train = sm.add_constant(X_train) # must have this line.
X_opt = X_train[:, [0, 1, 2, 3, 4,5]] # why need to write 0~5 (put all data into X_opt)
X_opt = np.array(X_opt, dtype='float') # change X_opt to np.array.


regressor_OLS = sm.OLS(y_train, X_opt).fit()
print(regressor_OLS.summary())
```

```python
X_opt = X_train[:, [0, 1, 3, 4, 5]] # 因為上面的x2的P VALUE > 0.05，所以我們就把2丟掉
X_opt = np.array(X_opt, dtype='float')
regressor_OLS = sm.OLS(endog = y_train, exog = X_opt).fit()
print(regressor_OLS.summary())
```

```python
X_opt = X_train[:, [0, 1, 3,  5]] # 因為上面的4的P VALUE > 0.05，所以我們就把4丟掉
X_opt = np.array(X_opt, dtype='float')
regressor_OLS = sm.OLS(endog = y_train, exog = X_opt).fit()
print(regressor_OLS.summary())
```

```python
X_opt = X_train[:, [0, 1, 3]] # 因為上面的5的P VALUE > 0.05，所以我們就把5丟掉
X_opt = np.array(X_opt, dtype='float')
regressor_OLS = sm.OLS(endog = y_train, exog = X_opt).fit()
print(regressor_OLS.summary())
```
